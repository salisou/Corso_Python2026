### impartazione librerie

```python
import pyodbc as odbc
import pandas as pd
```

Connessione al db 

```python
conn = odbc.connect(
    r"""
    Driver={ODBC Driver 17 for SQL Server};
    Server=(localdb)\P;
    Database=ScuolaDb;
    Trusted_Connection=yes;
    TrustServerCertificate=yes;
    """
)

print("Connessione Ok!")
```

Recuperazione delle tabelle

```python
q = """
    SELECT TABLE_NAME
    FROM INFORMATION_SCHEMA.TABLES
    WHERE TABLE_TYPE = 'BASE TABLE'
"""

df_tables = pd.read_sql(q, conn)

# lista delle tabelle 
tabelle = df_tables["TABLE_NAME"].tolist()

print("🗄️ Tabelle trovate:", tabelle)
```

Exportazione su Excel

```python
# crea una stringa con il nome del file 
file_excel = "Backup_automatico_scuolaDb1.xlsx"

with pd.ExcelWriter(file_excel, engine="openpyxl") as writer:
   for tabella in tabelle:
        print(f"📤 Esporto: {tabella}")
        
        # legge tutte le tabelle
        df = pd.read_sql(f"SELECT * FROM {tabella}", conn)
        
        # Scrive ogni tabella in un foglio Excel
        df.to_excel(writer, sheet_name=tabella, index=False)
conn.close()

print("\n🎉 BACKUP COMPLETTATO CON SUCCESSO!")
print(f"File creato 📃 {file_excel}")
```

```python
# crea una stringa con il nome del file 
file_excel = "Backup_scuolaDB.xlsx"

with pd.ExcelWriter(file_excel, engine="openpyxl") as writer:
  for tabella in tabelle:
        count = pd.read_sql(f"SELECT COUNT(*) AS Totale FROM {tabella}", conn)
        print(f"{tabella}: {count['Totale'][0]} record")
        
        # legge tutte le tabelle
        df = pd.read_sql(f"SELECT * FROM {tabella}", conn)
        
        # Scrive ogni tabella in un foglio Excel
        df.to_excel(writer, sheet_name=tabella, index=False)
conn.close()

print("\n🎉 BACKUP COMPLETTATO CON SUCCESSO!")
print(f"File creato 📃 {file_excel}")
```
