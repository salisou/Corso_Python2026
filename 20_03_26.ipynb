Connezione al server SQL Server 

```python
import pyodbc as odbc
import pandas as pd

conn = odbc.connect(
    r"""
    Driver={ODBC Driver 17 for SQL Server};
    Server=(localdb)\P;
    Database=ScuolaDb;
    Trusted_Connection=yes;
    TrustServerCertificate=yes;
    """
)

print("✅ Connessione riuscita!")
```

Visualizzazione della lista degli studenti

```python

df = pd.read_sql("EXEC sp_GetAllStudenti", conn)
df
```

Salva in excel

```python
df.to_excel("studenti.xlsx", index=False)
print("📂 File Excel creato con successo!")
```
