```python
import pyodbc as odbc
import pandas as pd
from datetime import datetime

conn = odbc.connect(
    r"""
    Driver={ODBC Driver 17 for SQL Server};
    Server=(localdb)\P;
    Database=ScuolaDb;
    Trusted_Connection=yes;
    TrustServerCertificate=yes;
    """
)

print("Connessione Ok!")
```

#### Recupera Tabelle

```python
query = """
    SELECT TABLE_NAME
    FROM INFORMATION_SCHEMA.TABLES
    WHERE TABLE_TYPE = 'BASE TABLE'
    AND TABLE_SCHEMA = 'dbo'
    ORDER BY TABLE_NAME
"""


df_tables = pd.read_sql(query, conn)

# lista delle tabelle 
tabelle = df_tables["TABLE_NAME"].tolist()

print("🗄️ Tabelle trovate:", tabelle)
```

Nome File Con Data 

```python

data_oggi = datetime.now().strftime("%Y-%m-%d_%H-%M")

# crea una stringa con il nome del file 
file_excel = f"Backup_Scuola_Db_{data_oggi}.xlsx"
```

Lista Statiche

```python
stats = []
```

Export 

```python
with pd.ExcelWriter(file_excel, engine="openpyxl") as writer:
   for tabella in tabelle:
        print(f"📤 Esporto: {tabella}")
        
        # legge tutte le tabelle
        df = pd.read_sql(f"SELECT * FROM {tabella}", conn)
        
        # Scrive ogni tabella in un foglio Excel
        df.to_excel(writer, sheet_name=tabella, index=False)
        
        # Conteggio del record
        totale = len(df)
        
        # Numoro Colonne
        colonne = len(df.columns)
        
        # Salva Statiche
        stats.append(
            {
                "Tabella": tabella,
                "Numero Record": totale,
                "Nomero Colonne": colonne
            }
        )
        # Foglio Riepilogo
        df_stats = pd.DataFrame(stats)
        
        df.to_excel(writer, sheet_name="Riepilo", index=False)
        
# chiusura 
conn.close()

print("\n🎉 BACKUP COMPLETTATO CON SUCCESSO!")
print(f"File creato 📃 {file_excel}")
```
