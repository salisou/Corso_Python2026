```python
class Studente:
    def __init__(self, nome, cognome, email, data_nascita, codice_fiscale): 
        self._nome = nome 
        self._cognome = cognome
        self._email = email
        self._data_nascita = data_nascita
        self._codice_fiscale = codice_fiscale
        
    def __str__(self):
        return f"{self._nome}, {self._cognome}, {self._email}, {self._data_nascita}, {self._codice_fiscale}"
    
    def mostra_dati(self):
        print(f"{self._nome}, {self._cognome}, {self._email}")
    
 
```

```python
class RegistroStudenti:
    def __init__(self):
        self._ListaStudenti = []
    
    def aggiungi_studente(self, studente):
        self._ListaStudenti.append(studente)
        
    def mostra_studenti(self):
        for s in self._ListaStudenti:
            print(s)
    
    def cerca_per_email(self, email):
        for st in self._ListaStudenti:
            if st._email == email:
                return st
        return None
    
    def cerca_per_nme(self, nome):
        for s in self._ListaStudenti:
            if s._nome == nome:
                return s
        return None


    def modifica_email(self, vecchio_email, nuovo_email):
        stud = self.cerca_per_email(vecchio_email)
        if stud:
            stud._email = nuovo_email
            return True
        return False

    def elimina_studente(self, email):
        for stu in self._ListaStudenti:
            if stu._email == email:
                self._ListaStudenti.remove(stu)
                return stu
        return False
```

## Menu 

```python
def menu():
    print("\n========= MENU ============")
    print("1. Aggiungi Studente")
    print("2. Mostra Studenti")
    print("3. Cerca Studente")
    print("4. Modifica Studente")
    print("5. Elimina Studente")
    print("6. Esci")


def verifica_ricerca(email):
    s = registro.cerca_per_email(email)
    if s:
        print("Studente trovato:")
        print(s)
    else:
        print("Studente non trovato\n")


def student_scelto():
    nome = input("Nome: ")
    cognome = input("Cognome: ")
    email = input("Email: ")
    data = input("Data di nascita: ")
    cf = input("Codice Fiscale: ")
    return nome, cognome, email, data, cf


def modifica_studente():
    vecchiEmail = input("Email dello studente da modificare: ")
    nuovoEmail = input("Nuova email: ")
        
    if registro.modifica_email(vecchiEmail, nuovoEmail):
        print("Email aggiornata 📧")
        
        studente = registro.cerca_per_email(nuovoEmail)
        if studente:
            print("Studente aggiornato:")
            print(studente)
        else:
            print("Errore: studente non trovato dopo modifica")
    else:
        print("Aggiornamento fallito")

    input("\nPremi INVIO per tornare al menu...")


def elimina_studente():
    email = input("Email da eliminare: ")
        
    if registro.elimina_studente(email):
        print("Studente eliminato con successo")
    else:
        print("Email non trovata")
            
    registro.mostra_studenti()
    input("\nPremi INVIO per tornare al menu...")


# 🔁 CICLO PRINCIPALE
while True:
    menu()
    
    scelta = input("Scegli un'opzione: ")
    
    if scelta == "1":
        nome, cognome, email, data, cf = student_scelto()
        
        s = Studente(nome, cognome, email, data, cf)
        registro.aggiungi_studente(s)
        print("\nStudente salvato con successo!")
        input("\nPremi INVIO per continuare...")

    elif scelta == "2":
        registro.mostra_studenti()
        input("\nPremi INVIO per continuare...")

    elif scelta == "3":
        email = input("Inserisci la mail: ")
        verifica_ricerca(email)
        input("\nPremi INVIO per continuare...")

    elif scelta == "4":
        modifica_studente()

    elif scelta == "5":
        elimina_studente()
        
    elif scelta == "6":
        print("Fine del programma...")
        break

    else:
        print("Scelta non valida!")
        input("\nPremi INVIO per continuare...")
```
