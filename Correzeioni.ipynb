## Soluzione Esercizio 1
##### Creare un istogramma dei voti

```python
import pandas as pd
import matplotlib.pyplot as plt

voti = pd.read_csv("voti.csv")
```

```python
voti["voto"].plot(kind="hist")

plt.title("Distribuzione dei voti")

plt.xlabel("Voto")

plt.ylabel("Numero studenti")

plt.show()
```

<h3>Soluzione Esercizio 2</h3>
<p>Creare un grafico a barre della media voti per materia <strng>Prima calcoliamo la media.</strng> </p>

```python
media_materie = voti.groupby("materiaId")["voto"].mean()
```

Poi il grafico

```python
media_materie.plot(kind="bar")

plt.title("Media voti per materia")

plt.xlabel("Materia")

plt.ylabel("Media voto")

plt.show()
```

<h3>Soluzione Esercizio 3</h3>
<p>Creare un grafico a torta dei voti <strong>Prima contiamo quanti voti ci sono.</strong></p>

```python
conteggio_voti = voti["voto"].value_counts()
```

poi il grafico

```python
conteggio_voti.plot(kind="pie", autopct="%1.1f%%")

plt.title("Distribuzione dei voti")

plt.show()
```
