```python
import pandas as pd
import matplotlib.pyplot as plt
```

Leggere il CSV

```python
df = pd.read_csv("studente.csv")
```

Grafico a barre dei voti

```python
df.plot(x="nome", y="voto", kind="bar")

plt.show()
```

Istogramma dei voti

Serve per vedere la distribuzione dei voti.

```python
df["voto"].plot(kind="hist")

plt.show()
```

Contare studenti per corso

Prima contiamo:

```python
conteggio = df["corso"].value_counts()
```

Poi Grafico

```python
conteggio.plot(kind="bar")

plt.show()
```

Grafico a torta

Molto intuitivo per gli studenti.

```python
df["corso"].value_counts().plot(kind="pie")

plt.show()
```

Migliorare il grafico

Possiamo aggiungere titolo e etichette.

```python
df.plot(x="nome", y="voto", kind="bar")

plt.title("Voti degli studenti")
plt.xlabel("Studente")
plt.ylabel("Voto")

plt.show()
```
