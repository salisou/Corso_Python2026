Letture del file ("r")

```python
import os 

nome_file = "esempio_uno.text"

if os.path.exists(nome_file):
    print(f"il file {nome_file} esite.")
else:
    print(f"il file {nome_file} non esite.")
```

## Lettura dei contenuti di un file

```python
file = open(nome_file, "r") 
contenuto = file.read() # Letture del contenuto del file

file.close()

print(contenuto) # Stampa il contenuto del file
```

<h3>Lettura linea per linea</h3>
<p>è possibile leggere un file linea per linea utilizzando iun ciclo <i>for</i></p>

```python
mio_File =  "esempio_uno.text"

with open(mio_File, "r") as file:
    for linea in file:
        print(linea.split()) #.split() rimuove il carettere newline (\n)
```

<h3>Scrittura in un file</h3>

```python
__file = open("esempio_uno.text", "a")
__file.write("Nuovo contenuto del file.")
__file.close()
```

Aggiunta di Contenuti a un file 

```python
file01 = open("esempio_uno.text", "a")
file01.write(" Questo contenuto è un nuovo aggiornamento")
file01.close()
```
